```python
import numpy as np

import matplotlib.pyplot as plt

# Identification des colonnes numériques
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Détection des outliers avec la méthode IQR
for column in numeric_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Outliers dans la colonne {column} :")
    print(outliers[[column]])

# Visualisation des outliers avec des boxplots
df[numeric_columns].boxplot(figsize=(15, 7))
plt.xticks(rotation=45)
plt.title("Boxplots des colonnes numériques")
plt.show()
```
